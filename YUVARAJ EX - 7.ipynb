```python
import numpy as np

# Define the FIND-S algorithm
def find_s_algorithm(examples):
    # Initialize the hypothesis to the most specific hypothesis
    hypothesis = [None] * (len(examples[0]) - 1)

    for example in examples:
        if example[0] == 'Sunny':
            for i in range(len(hypothesis)):
                if hypothesis[i] is None:
                    hypothesis[i] = example[i]
                elif hypothesis[i] != example[i]:
                    hypothesis[i] = '?'

    return hypothesis


# Example training data
training_data = [
    ['Sunny', 'Warm', 'Normal', 'Strong', 'Warm', 'Same', 'Yes'],
    ['Sunny', 'Warm', 'High', 'Strong', 'Warm', 'Same', 'Yes'],
    ['Rainy', 'Cold', 'High', 'Strong', 'Warm', 'Change', 'No'],
    ['Sunny', 'Warm', 'High', 'Strong', 'Cool', 'Change', 'Yes']
]

# Run the FIND-S algorithm
hypothesis = find_s_algorithm(training_data)

# Print the hypothesis
print("The most specific hypothesis is:", hypothesis)
```
